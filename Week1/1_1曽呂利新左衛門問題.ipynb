# 【問題1】100日目の米粒の数
100日目には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）

```python
rice = 1
days = 100
one_day_rice_list = [rice]
total_rice_list = [rice]
for _ in range(days):
    rice *= 2
    one_day_rice_list.append(rice)
    total_rice_list.append(sum(one_day_rice_list))

print(f'100日目までに貰った米粒の合計：{total_rice_list[-1]}粒')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.title('one_day_rice')
plt.xlabel('days')
plt.ylabel('amount')
plt.plot(one_day_rice_list)
plt.show()

plt.title('total_rice')
plt.xlabel('days')
plt.ylabel('amount')
plt.plot(total_rice_list)
plt.show()
```

# 【問題2】100日目以外の米粒の数
この逸話は30日だったり81日だったりと日数が曖昧です。どの話にも対応できるように日数を引数とした 関数 を作成しましょう。関数を作成した上で、コースを受講する日数の間でいくつ米粒をもらえるかを答えてください。前の問題同様に折れ線グラフも出力してください。

```python
def compute_sorori_shinzaemon(day=100):
    rice = 1
    list_n_grains = [rice]
    list_total_grains = [rice]
    for _ in range(day):
        rice *= 2
        list_n_grains.append(rice)
        list_total_grains.append(sum(list_n_grains))
    
    return list_n_grains, list_total_grains
```

```python
# 受講期間は、120日とする。
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=120)
total = list_total_grains[-1]
print(f'受講中に貰った米粒の合計：{total}粒')
```

```python
plt.title('list_n_grains')
plt.xlabel('days')
plt.ylabel('amount')
plt.plot(list_n_grains)
plt.show()

plt.title('list_total_grains')
plt.xlabel('days')
plt.ylabel('amount')
plt.plot(list_total_grains)
plt.show()
```

# 【問題3】何人が何日生活できる米か
この米があれば何人が何日生活できるか計算する関数を作成してください。コースを受講する期間に貰える米粒で、このクラスにいる人数であれば、何日間生活できるかを出力してみましょう。

関数の最低限の入力は 米粒の数(int) と 人数(int) 、出力は 生活できる日数(int) です。関数名やコメントなどは分かりやすく考えて作成してください。

ここでは1日に必要な米粒の数は提示しません。自分なりに調べて、計算式に落とし込んでください。どう調べ、何を根拠にしたかもコメントを残しておいてください。

```python
def live_days(rice_number, count, required_number):
    return rice_number // (count*required_number)
```

【生存日数を求めるのに必要な値】
- 米粒の数：total (2658455991569831745807614120560689151粒)  
- 人数：12人
- １人あたり1日に必要な米粒の数：12000粒  

【必要な米粒の数の根拠】  
お米1000粒を炊くと約20g位の重さになる。  
なので、1合(150g)のお米だと、7500～8000粒くらい。  
1合のごはんをお茶碗に盛ると、だいたいお茶碗２杯分になるので、  
茶碗一杯、約4000粒と推測できる。  
それを1日3食として、1日に必要な米粒の数を求めた。

```python
print(f'{live_days(total, 12, 12000)}日生活できる。')
```
