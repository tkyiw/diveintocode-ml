# 【問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

$$
\begin{bmatrix}
1.0 & 0.8 \\
0.8 & 1.0 \\
\end{bmatrix}
$$

多次元の正規分布による乱数はnp.random.multivariate_normal()を使うことで作成できます。作成したndarrayのshapeは(500, 2)になります。

《シード値について》

実行ごとに値が変わらないように、np.random.seed()により乱数のシード値を固定しておきましょう。今回シード値は0を使うことにします。

```python
import numpy as np

mu = [-3, 0]
sigma = [[1.0, 0.8], [0.8, 1.0]]
 
# 2次元正規乱数を500個生成
np.random.seed(0)
values1 = np.random.multivariate_normal(mu, sigma, 500)
```

```python
values1.shape
```

```python
values1
```

# 【問題2】散布図による可視化
問題1で作成したデータ点を散布図により可視化してください。

散布図はmatplotlibのplt.scatter()を使うことで描けます。

```python
import matplotlib.pyplot as plt
plt.title('scatter')
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter(values1[:,0], values1[:,1], alpha=0.3);
```

# 【問題3】ヒストグラムによる可視化
問題1で作成したデータをヒストグラムにより可視化してください。

ヒストグラムはplt.hist()を使うことで描けます。

各次元ごとに作成し、2つのヒストグラムとしてください。

それぞれのヒストグラムの横軸の範囲はplt.xlim()を用いて揃えるようにしてください。

```python
plt.hist(values1[:,0], bins=30)
plt.xlim(-6, 3);
```

```python
plt.hist(values1[:,1], bins=30)
plt.xlim(-6, 3);
```

# 【問題4】データの追加
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

$$
\begin{bmatrix}

\end{bmatrix}
$$

```python
mu2 = [0, -3]
sigma = [[1.0, 0.8], [0.8, 1.0]]
 
# 2次元正規乱数を500個生成
np.random.seed(0)
values2 = np.random.multivariate_normal(mu2, sigma, 500)
```
